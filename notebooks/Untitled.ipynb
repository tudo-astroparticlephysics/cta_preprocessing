```python
import fact.io

from scipy.stats import binned_statistic
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename='./old_method/gamma_20deg_0deg_run501___cta-prod3-merged_desert-2150m-Paranal-3HB89-NGFD.hdf5'
df = fact.io.read_data(filename, key='array_events')
df.columns
```

```python

distance = np.sqrt((df.mc_core_x - df.core_x_prediction)**2 + (df.mc_core_y - df.core_y_prediction)**2)

plt.scatter(np.log10(df.mc_energy), distance)
plt.yscale('log')

b_50, bin_edges, binnumber = binned_statistic(np.log10(df.mc_energy), distance, statistic=np.nanmedian, bins=10)

plt.plot(0.5 * (bin_edges[:-1] + bin_edges[1:]), b_50, color='xkcd:blue', lw=3)

plt.ylabel('distance to true impact / m')
plt.xlabel('log10(energy/TeV)')
plt.savefig('impact_compare.png')

bin_edges, b_50
```
